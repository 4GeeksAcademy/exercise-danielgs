#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
# TODO

import pandas as pd
ds = pd.read_csv('assets/real_estate.csv', sep=';')

cinturon_sur = ds[ds['level5'].isin(["Fuenlabrada","Leganés","Getafe","Alcorcón"])]
cinturon_sur

import matplotlib.pyplot as plt

for poblacion in ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]:
    precios_normalizados = (cinturon_sur[cinturon_sur['level5'] == poblacion]['price'] - cinturon_sur[cinturon_sur['level5'] == poblacion]['price'].min()) / (cinturon_sur[cinturon_sur['level5'] == poblacion]['price'].max() - cinturon_sur[cinturon_sur['level5'] == poblacion]['price'].min())
    plt.hist(precios_normalizados, bins=20, alpha=0.5, label=poblacion)

plt.title('Distribución normalizada de precios por población')
plt.xlabel('Precio normalizado')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
# TODO
import pandas as pd
ds = pd.read_csv('assets/real_estate.csv', sep=';')

cinturon_sur = ds[ds['level5'].isin(["Fuenlabrada","Leganés","Getafe","Alcorcón"])]
cinturon_sur
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

poblaciones = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

for i, poblacion in enumerate(poblaciones):
    ax = axs[i//2, i%2]
    data = cinturon_sur[cinturon_sur['level5'] == poblacion]
    ax.scatter(data['m2'], data['price'], alpha=0.5)
    ax.set_title(poblacion)
    ax.set_xlabel('Superficie (m2)')
    ax.set_ylabel('Precio')

plt.tight_layout()
plt.show()
```
